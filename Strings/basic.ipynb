```python
#largest odd number in a string:
#brute force:
#Algorithm:
# 1. Initialize result = ""
# 2. Let n = length of string
# 3. Loop i from 0 to n-1:
#     a. Loop j from i to n-1:
#         i. Extract substring sub = num[i:j+1]
#         ii. Check if last digit of sub is odd:
#             - If odd:
#                 → If length of sub > length of result:
#                         update result = sub
#                 → Else if lengths equal and sub > result:
#                         update result = sub
# 4. Return result

# def largestOdd(num):
#     result = ""
#     for i in range(len(num)):
#         for j in range(i,len(num)):
#             sub = num[i:j+1]
#             if int(sub[-1])%2 != 0:
#                 if (len(sub) > len(result)) or (len(sub) == len(result) and sub > result):
#                     result = sub
#     return result

# num = input()
# print(largestOdd(num))
#Time complexity = O(n**3)
#Space complexity = O(n)

#optimal:
# Algorithm:
# 1. Let n = length of string

# 2. Loop i from n-1 to 0:
#     a. Check if num[i] is odd:
#         → If yes:
#             return substring num[0:i+1]

# 3. If no odd digit found:
#     → return ""

def largestOdd(num):
    for i in range(len(num)-1,-1,-1):
        if int(num[i]) % 2 != 0:
            return num[:i+1]
    return ""

num = input()
print(largestOdd(num))
#Time complexity = O(n)
#Space complexity = O(1)
```

```python
#Remove the outermost paranthesis:
#brute force:
# Algorithm:
# 1. Initialize:
#    result = empty string
#    count = 0
#    start = 0
# 2. Traverse the string from left to right
# 3. For each character:
#    - If '(' → increase count
#    - If ')' → decrease count
# 4. Whenever count becomes 0:
#    - We found one primitive substring
#    - Extract substring from start to current index
#    - Remove its first and last characters
#    - Add remaining part to result
#    - Update start = current index + 1
# 5. Return result

# def remove(s):
#     result = ""
#     count = 0
#     start = 0
#     for p in range(len(s)):
#         if s[p] == "(":
#             count += 1
#         else:
#             count -= 1
#         if count == 0:
#             prem = s[start:p+1]
#             result += prem[1:-1]
#             start = p+1
#     return result

# s = input()
# print(remove(s))
#Time complexity = O(n)
#Space complexity = O(n)

#optimal:
#Algorithm:
# 1. Take an empty result string.
# 2. Maintain a variable "depth" to track nesting level.
#    Initially, depth = 0.
# 3. Traverse the string character by character.
# 4. For each character:
#    CASE 1: If it is '('
#       - If current depth is greater than 0,
#         then this is NOT an outer parenthesis → include it in result.
#       - Increase depth by 1.
#    CASE 2: If it is ')'
#       - Decrease depth by 1.
#       - If updated depth is still greater than 0,
#         then this is NOT an outer parenthesis → include it in result.
# 5. Continue until the string ends.
# 6. Return the result.

# core idea:
# - Outer '(' appears when depth = 0 → skip it
# - Outer ')' appears when depth becomes 0 → skip it
# - All inner parentheses → keep them

def remove(s):
    result = ""
    count = 0
    for p in s:
        if p == "(":
            if count > 0:
                result += p
            count += 1
        else:
            if count > 1:
                result += p
            count -= 1
    return result

s = input()
print(remove(s))
#Time complexity = O(n)
#Space complexity = O(1)
```

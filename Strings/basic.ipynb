```python
#Remove the outermost paranthesis:
#brute force:
# Algorithm:
# 1. Initialize:
#    result = empty string
#    count = 0
#    start = 0
# 2. Traverse the string from left to right
# 3. For each character:
#    - If '(' → increase count
#    - If ')' → decrease count
# 4. Whenever count becomes 0:
#    - We found one primitive substring
#    - Extract substring from start to current index
#    - Remove its first and last characters
#    - Add remaining part to result
#    - Update start = current index + 1
# 5. Return result

# def remove(s):
#     result = ""
#     count = 0
#     start = 0
#     for p in range(len(s)):
#         if s[p] == "(":
#             count += 1
#         else:
#             count -= 1
#         if count == 0:
#             prem = s[start:p+1]
#             result += prem[1:-1]
#             start = p+1
#     return result

# s = input()
# print(remove(s))
#Time complexity = O(n)
#Space complexity = O(n)

#optimal:
#Algorithm:
# 1. Take an empty result string.
# 2. Maintain a variable "depth" to track nesting level.
#    Initially, depth = 0.
# 3. Traverse the string character by character.
# 4. For each character:
#    CASE 1: If it is '('
#       - If current depth is greater than 0,
#         then this is NOT an outer parenthesis → include it in result.
#       - Increase depth by 1.
#    CASE 2: If it is ')'
#       - Decrease depth by 1.
#       - If updated depth is still greater than 0,
#         then this is NOT an outer parenthesis → include it in result.
# 5. Continue until the string ends.
# 6. Return the result.

# core idea:
# - Outer '(' appears when depth = 0 → skip it
# - Outer ')' appears when depth becomes 0 → skip it
# - All inner parentheses → keep them

def remove(s):
    result = ""
    count = 0
    for p in s:
        if p == "(":
            if count > 0:
                result += p
            count += 1
        else:
            if count > 1:
                result += p
            count -= 1
    return result

s = input()
print(remove(s))
#Time complexity = O(n)
#Space complexity = O(1)
```

```python
#longest common prefix:
#brute force:
#Algorithm:
# 1. If the array of strings is empty, return ""
# 2. Take the first string as reference
#    first = strs[0]
# 3. Initialize an empty string:
#    prefix = ""
# 4. For each index i from 0 to length of first string:
#    a. Let ch = first[i]
#    b. For each string s in strs starting from index 1:
#       i. If i >= length of s OR s[i] != ch:
#             return prefix
#    c. If all strings have same character:
#          prefix = prefix + ch
# 5. Return prefix

# def longestcommonprefix(strs):
#     if not strs:
#         return ""
#     result = ""
#     first = strs[0]
#     for i in range(len(first)):
#         ch = first[i]
#         for s in strs[1:]:
#             if i >= len(s) or s[i] != ch:
#                 return result
#         result += ch
#     return result
#Time complexity = O(n*m)
#Space complexity = O(1)

#optimized:
#Algorithm:
# 1. If the array of strings is empty, return ""
# 2. Sort the array of strings
# 3. Take:
#    first = strs[0]
#    last = strs[n - 1]
# 4. Initialize i = 0
# 5. While i < length of first AND i < length of last AND first[i] == last[i]:
#       i = i + 1
# 6. Return substring of first from index 0 to i

def longestcommonprefix(strs):
    if not strs:
        return ""
    strs.sort()
    first = strs[0]
    last = strs[-1]
    i = 0
    while i < len(first) and i < len(last) and first[i] == last[i]:
        i += 1
    return first[:i]
#Time complexity = O(nlogn + m)
#Space complexity = O(1)

strs = list(map(str,input().split()))
print(longestcommonprefix(strs))
```

```python
#isomorphic string:
#brute force:
# Algorithm: 
# Brute Force (Isomorphic Strings)
# 1. If length of s != length of t:
#       return False
# 2. For i from 0 to n-1:
#       c1 = s[i]
#       c2 = t[i]
# 3. For j from 0 to i-1:
#       If s[j] == c1 and t[j] != c2:
#             return False
#       If t[j] == c2 and s[j] != c1:
#             return False
# 4. If no conflicts found:
#       return True

# def isomorphic(s,t):
#     if len(s) != len(t):
#         return False
#     for i in range(len(s)):
#         c1 = s[i]
#         c2 = t[i]
#         for j in range(0,i):
#             if s[j] == c1 and t[j] != c2:
#                 return False
#             if t[j] == c2 and s[i] != c1:
#                 return False
#     return True
#Time complexity = O(n**2)
#Space complexity = O(1)

#optimal:
# Algorithm: 
# Optimal (Using Hashmaps)
# 1. If length of s != length of t:
#       return False
# 2. Create two hashmaps:
#       map_s_t = {}
#       map_t_s = {}
# 3. For i from 0 to n-1:
#       c1 = s[i]
#       c2 = t[i]
# 4. If c1 exists in map_s_t:
#       If map_s_t[c1] != c2:
#             return False
#    Else:
#       map_s_t[c1] = c2
# 5. If c2 exists in map_t_s:
#       If map_t_s[c2] != c1:
#             return False
#    Else:
#       map_t_s[c2] = c1
# 6. If loop completes:
#       return True

def isomorphic(s,t):
    if len(s) != len(t):
        return False
    map_s_t = {}
    map_t_s = {}
    for i in range(len(s)):
        c1 = s[i]
        c2 = t[i]
        if c1 in map_s_t:
            if map_s_t[c1] != c2:
                return False
        else:
            map_s_t[c1] = c2
        if c2 in map_t_s:
            if map_t_s[c2] != c1:
                return False
        else:
            map_t_s[c2] = c1
    return True
#Time complexity = O(n)
#Space complexity = O(n)

s = input()
t = input()
print(isomorphic(s,t))
```

```python
#Rotate String:
#brute force:
# Algorithm: 
# Brute Force (Rotate String)
# 1. If length of s != length of goal:
#       return False
# 2. Repeat n times (where n = length of s):
#       a. If s == goal:
#             return True
#       b. Rotate string s by 1 position:
#             s = s[1:] + s[0]
# 3. If no match found after all rotations:
#       return False

# def Rotate(s,goal):
#     if len(s) != len(goal):
#         return False
#     for i in range(len(s)):
#         if s == goal:
#             return True
#         s = s[1:] + s[0]
#     return False
#Time complexity = O(n**2)
#Space complexity = O(n)

#optimized:
# Algorithm: 
# Optimal (Using String Concatenation)
# 1. If length of s != length of goal:
#       return False
# 2. Create a new string:
#       temp = s + s
# 3. Check:
#       If goal is a substring of temp:
#             return True
#       Else:
#             return False

def Rotate(s,goal):
    if len(s) != len(goal):
        return False 
    return goal in s+s
#Time complexity = O(n)
#Space complexity = O(n)

s = input()
goal = input()
print(Rotate(s,goal))
```

```python
# largest element in an array :
#Brute force:
#1. sort the arr in ascending order.
#2. return last element of the array.

# def largest(nums):
#     for i in range(len(nums)):
#         for j in range(0,len(nums)-i-1):
#             if nums[j] > nums[j+1]:
#                 nums[j],nums[j+1] = nums[j+1],nums[j]
#     return nums[len(nums)-1]

# nums = list(map(int,input("Enter array Elements: ").split()))
# print(largest(nums))
#Time complexity = O(n**2)
#Space complexity = O(1)


# optimal approach:
 
# 1.Create a variable called max and initialize it with the value of the first element in the array.
# 2.Use a for loop to iterate through the rest of the elements in the array.
# 3.In each iteration, compare the current element with the max variable.
# 4.If the current element is greater than the max value, update the max value with the current element's value.
# 5.After completing the loop, print the max variable, which will hold the largest value in the array.

def largest(nums):
    large = nums[0]
    for i in range(len(nums)):
        if nums[i] > large:
            large = nums[i]
    return large

nums = list(map(int,input("Enter array Elements: ").split()))
print(largest(nums))
#Time complexity = O(n)
#Space complexity = O(1)
```

```python
#second smallest and second largest element in the arr:
#Brute force approach:

#Algorithm: Find the Second Largest Element in an Array (Brute Force)
# Step 1: Sort the array in ascending order.
# Step 2: Check the length of the array.if the length of the array is 1, return arr[0],arr[0].
# Step 3: If the length of the array is 2, return arr[0] and arr[1].
# Step 4: Convert the array into a set to remove duplicate elements.arr = sorted(set(arr))
# Step 5: Check the length of the array after removing duplicates.If the length is 1, return arr[0] and arr[0].
# Step 6: Otherwise, return arr[-2] as the second largest element.

# def second_largest_smallest(nums):
#     nums.sort()
#     if len(nums) == 1:
#         return nums[0] , nums[0]
#     elif len(nums) == 2:
#         return nums[0] , nums[1]
#     else:
#         nums = sorted(set(nums))
#         if len(nums) == 1:
#             return nums[0],nums[0]
#         return nums[1], nums[-1]
# nums = list(map(int,input("Enter array Elements: ").split()))
# print(second_largest_smallest(nums))
#Time complexity = O(nlogn)
#Space complexity = O(n)

#optimized approach:
#Algorithm:
# 1.We will need four variables: small, second_small, large, and second_large. Initialize small and second_small to inf, and large and second_large to -inf.
# 2.Second Smallest Algorithm: 
# 2.1 If the current element is smaller than 'small', update the values of second_small and small.
# 2.2 Else if the current element is smaller than 'second_small', update the value of second_small.
# 2.3 After traversing the array, the second smallest element will be stored in the variable second_small.
# 3. Second Largest Algorithm:
# 3.1 If the current element is larger than 'large', update the values of second_large and large.
# 3.2 Else if the current element is larger than 'second_large', update the value of second_large.
# 3.3 After traversing the array, the second largest element will be stored in the variable second_large.
def second_largest_smallest(nums):
    small = float('inf')
    second_small = float('inf')
    large = float('-inf')
    second_large = float('-inf')

    for num in nums:
        if num < small:
            second_small = small
            small = num
        elif num < second_small and num != small:
            second_small = num
        if num > large:
            second_large = large
            large = num
        elif num > second_large and num != large:
            second_large = num
    return second_small,second_large

nums = list(map(int,input("Enter array Elements: ").split()))
print(second_largest_smallest(nums))
#Time complexity = O(n)
#Space complexity = O(1)
```

```python
#check if array is sorted II:
#Brute force:
#Algorithm:
# Check if Array is Sorted and Rotated (Brute Force)
# Step 1: Read the array nums.
# Step 2: Let n be the length of the array.
# Step 3: Repeat the following steps for each possible rotation from 0 to n−1:
#     a) Assume the array is sorted.
#        Set is_sorted = True.
#     b) Traverse the array from index i = 0 to n−2.
#     c) If nums[i] > nums[i+1], then the array is not sorted.
#        Set is_sorted = False and break the loop.
#     d) If is_sorted remains True after the traversal,
#        return True because a sorted rotation exists.
#     e) Rotate the array by one position to the right:
#        nums = [nums[-1]] + nums[:-1]
# Step 4: If no rotation produces a sorted array,
# return False.

# def check_sorted(nums):
#     for rotation in range(len(nums)):
#         sorted = True
#         for i in range(len(nums)-1):
#             if nums[i] > nums[i+1]:
#                 sorted = False
#                 break
#         if sorted:
#             return True
#         nums = [nums[-1]]+nums[:-1]
#     return False

# nums = list(map(int,input("Enter array Elements: ").split()))
# print(check_sorted(nums))

#optimized:
#algorithm:
# Check if an Array is Sorted and Rotated
# Step 1: Let n be the length of the array.
# Step 2: Initialize a variable count = 0.
# Step 3: Traverse the array from i = 0 to n-1.
# Step 4: For each index i, compare the current element with the next element using circular indexing:
# If nums[i] > nums[(i + 1) % n], increment count by 1.
# Step 5: After completing the traversal, check the value of count.
# Step 6: If count is less than or equal to 1, return True because the array is sorted and rotated.
# Step 7: Otherwise, return False because the array is not sorted and rotated.

def check_sorted(nums):
    drops = 0
    for i in range(len(nums)):
        if nums[i] > nums[(i+1)%len(nums)]:
            drops += 1
    if drops == 0 or drops == 1:
        return True
    else:
        return False  

nums = list(map(int,input("Enter array Elements: ").split()))
print(check_sorted(nums))
#Time complexity = O(n)
#Space complexity = O(1)
```

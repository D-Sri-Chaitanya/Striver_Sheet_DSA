```python
#Reverse Linked List
#brute force:
def reverse(head):
    arr = []
    curr = head
    while curr:
        arr.append(curr.val)
        curr = curr.next
    arr.reverse()
    curr = head
    i = 0
    while curr:
        curr.val = arr[i]
        i += 1
        curr = curr.next
    return head
```
